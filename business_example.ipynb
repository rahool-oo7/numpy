```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]\n",
sales_data = np.array([
        [1, 150000, 270000, 230000, 250000],  # Paradise Biryani
        [2, 120000, 140000, 160000, 190000],  # Beijing Bites
        [3, 200000, 230000, 260000, 300000],  # Pizza Hub\n
        [4, 270000, 210000, 240000, 270000],  # Burger Point\n
        [5, 160000, 185000, 205000, 230000]   # Chai Point\n
    ])
print("==== Swiggy Sales Analysis ====")

print("\nShape of data :", sales_data.shape)
```

```python
print("Sales data of first 3 restaurants : \n", sales_data[0:3])

#Total sales per year separte
print("Total sales of year 2021", np.sum(sales_data[:, 1]))
print("Total sales of year 2022", np.sum(sales_data[:, 2]))
print("Total sales of year 2023", np.sum(sales_data[:, 3]))
print("Total sales of year 2024", np.sum(sales_data[:, 4]))
```

```python
#Total sales per year combined
print("Yearly sales combined : ", np.sum(sales_data[:, 1:], axis=0))

#yearly combined sales of every company
print('yearly combined sales of each company : ', np.sum(sales_data[:, 1:], axis=1))
```

```python
#SUM of yearly combined sales of every company
yearly_sum = sum(np.sum(sales_data[:, 1:], axis=1))
print("SUM of yearly combined sales of every company : ", yearly_sum)
```

```python
#Minimum sales per company (axis=1)
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum sales per company : ", min_sales)

#Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum sales per year : ", max_sales)
```

```python
#Average sales per restaurant

# avg_sales = np.average(sales_data[:, 1:], axis=1) 
# ðŸ‘† this is correct but we could also use mean

avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per restaurant ; ", avg_sales)
```

```python
# Cumilative Sum (cumsum) per resaturant
cumisum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumisum)

#Visualize
plt.figure(figsize=(6, 4))
plt.plot(np.mean(cumisum, axis=0))
plt.title("Average cumilative sales across all restaurant")

years = [2021, 2022, 2023, 2024]
plt.xticks(ticks=range(len(years)), labels=years)

plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
cumisum = np.cumsum(sales_data[:, 1:], axis=1)

# Plot setup
years = [2020, 2021, 2022, 2023]
plt.figure(figsize=(8, 5))

# ðŸ”¹ Plot individual restaurant cumulative sales
for i in range(cumisum.shape[0]):
    plt.plot(cumisum[i], label=f"Restaurant {i+1}", linestyle='--', marker='o')

# ðŸ”¸ Plot average cumulative sales
avg_cumisum = np.mean(cumisum, axis=0)
plt.plot(avg_cumisum, label="Average", color="black", linewidth=2, marker='D')

# Custom ticks and labels
plt.xticks(ticks=range(len(years)), labels=years)

# Labels and legend
plt.title("Cumulative Sales of Restaurants (with Average)")
plt.xlabel("Year")
plt.ylabel("Cumulative Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
